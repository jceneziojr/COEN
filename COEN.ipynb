## Numpy

```python
import numpy as np

# Operações básicas com matrizes

A = np.array([
    [1, 2],
    [3, 4]
])

print(A)

B = np.array([
    [5, 6],
    [7, 8]
])

inversa_A = np.linalg.inv(A)
inversa_B = np.linalg.inv(B)

soma = A + B
subtracao = A - B
multiplicacao = np.matmul(A, B) #ou operador @
divisao = np.matmul(A, inversa_B)
transposta_A = A.T


print(f"Inversa de A:\n{inversa_A}")
print(f"\nInversa de B:\n{inversa_B}")

print(f"\nSoma:\n{soma}")
print(f"\nSubtração:\n{subtracao}")
print(f"\nMultiplicação:\n{multiplicacao}")
print(f"\nDivisão:\n{divisao}")
print(f"\nTransposta de A:\n{transposta_A}")
```

```python
#Resolvendo sistemas lineares do tipo Ax = B

A2 = np.array([
    [1, 2],
    [3, 4]
])

B2 = np.array([
    [0],
    [1],
])

sol = np.linalg.solve(A2, B2)
print(f"Solução do sistema linear:\n{sol}")
```

```python
# Inserção de elementos

A2 = np.vstack((A, [10,11]))

print(f"Exemplo de adição de linhas: \n{A2}")

A3 = np.hstack((A,[[12],[13]]))

print(f"\nExemplo de adição de colunas: \n{A3}")

print(f"\nDimensão da matriz A: {A.shape} / Dimensão da matriz A2: {A2.shape} / Dimensão da matriz A3: {A3.shape}\n")

# Remoção de linhas/colunas

A2 = np.delete(A2, 2, axis=0) # matriz, índice do elemento e dimensão a ser deletada (0 - linhas)

print(f"Matriz A2 após remover terceira linha: \n{A2}")

A3 = np.delete(A3, 2, axis=1) # matriz, índice do elemento e dimensão a ser deletada (1 - colunas)

print(f"\nMatriz A3 após remover terceira coluna: \n{A3}")

# Acessando elementos da matriz
# Primeiro índice em python é 0 SEMPRE

print(f"\nMatriz B elemento 1,1 -> {B[0][0]}")
print(f"Matriz A elemento 2,1 -> {A[1][0]}")
```

```python
# Criação e manipulação de arrays/vetores

x = np.linspace(0, 5, 6) # 10 valores no intervalo [0,5]
print(f"Exemplo linspace: vetor x -> {x}")

x1 = np.arange(0, 6, 0.75) # valores no intervalo [0,5), espaçados de 1
print(f"Exemplo arange: {x1}")

x2 = np.append(x, x1)
print(f"Exemplo de união de arrays: {x2}")

x3 = np.sort(x2)
print(f"Ordenando os valores da array: {x3}")

x4 = np.unique(x3)
print(f"Mantendo os valores exclusivos: {x4}")

x5 = np.delete(x, 3) # removendo o quarto elemento
print(f"Array x após retirar o quarto elemento ({x[3]}): {x5}")

x6 = np.delete(x, [-1, -2]) # removendo o último e o penúltimo elemento (usar "-" acessa os elementos de forma inversa)
print(f"Array x após retirar o último e o penúltimo elemento ({x[-1]}, {x[-2]}): {x6}")

x7 = np.delete(x, np.where(x==0))
print(f"Array x após retirar o(s) elemento(s) de valor 0: {x7}")

x8 = x[:3] #pegando um slice até o terceiro elemento (não incluso)
print(f"Slice do vetor x, até o terceiro elemento (não incluso): {x8}")

x9 = x[3:] #pegando um slice a partir do terceiro (não incluso)
print(f"Slice do vetor x, a partir o terceiro elemento (não incluso): {x9}")

print(f"Slice do vetor x, do segundo (não incluso) até o penúltimo elemento (não incluso): {x[2:-2]}")
```

```python
# Dimensões de arrays
print(f"Dimensão do vetor x: {x.shape}") #uma dimensão

x10 = np.reshape(x,(-1,1))
print(f"\nDimensão do vetor x após reshape: {x10.shape}")

x11 = np.reshape(x, (3,2))
print(f"\nVetor x convertido em uma matriz 3x2:\n{x11}")

x12 = x11.flatten()
print(f"\nMatriz x9 convertida de volta em um vetor: {x12}")
```

```python
# import scipy
from scipy import signal
```

Circuito RC

<img src="imagens/RC.png"> 

Função de Transferência: $H(s) = \frac{\frac{1}{RC}}{s+\frac{1}{RC}}$

```python
# Filtro LP RC (~50 Hz)

R = 100
C = 30e-6

num = [1/(R*C)]
den = [1, 1/(R*C)]

tf = signal.TransferFunction(num, den)

t = np.linspace(0, 2*np.pi, 10000)
input_sig = np.sin(2*np.pi*1*t) + np.sin(2*np.pi*1e3*t)

plt.plot(t,input_sig)

t_out, y_out, _ = signal.lsim(tf, input_sig, t)

plt.plot(t_out,y_out)
plt.title("Filtro RC")
plt.xlabel("Tempo")
plt.ylabel("Amplitude")
plt.tight_layout()
```

```python
# Análise FFT

input_sig = np.sin(2*np.pi*60*t) + np.sin(2*np.pi*100*t)

input_sig_fft = np.fft.fft(input_sig)
N = len(input_sig)
frequencies = np.fft.fftfreq(N, t[1] - t[0])

input_sig_mag = np.abs(input_sig_fft[:N//2])
frequencies = frequencies[:N//2]

plt.figure(figsize=(10, 6))
plt.plot(frequencies, input_sig_mag)
plt.title('FFT')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Magnitude')
plt.xscale("log")
```

Resposta ao degrau unitário: $H(s) = \frac{1}{s^2 + 2s + 1}$

```python
num = [1]
den = [1, 2, 1]

t = np.linspace(0, 10, 100)
input_sig = np.heaviside(t - 2, 1) # 0 enquanto t-2<0 e 1 quando t-2>0

tf = signal.TransferFunction(num, den)
t_out, y_out, _ = signal.lsim(tf, input_sig, t)

plt.plot(t, input_sig)
plt.plot(t_out, y_out)
plt.xlabel("Tempo")
plt.ylabel("Amplitude")
plt.title(r"Resposta ao Degrau unitário de $H(s) = \frac{1}{s^2 + 2s + 1}$")
```

```python
# Resposta ao Impulso

t_out, y_out = signal.impulse(tf, T=t)
plt.plot(t_out, y_out)
plt.xlabel("Tempo")
plt.ylabel("Amplitude")
plt.title(r"Resposta ao Impulso de $H(s) = \frac{1}{s^2 + 2s + 1}$")
```

```python
# Localização de picos

x = np.random.randn(100)
peaks, _ = signal.find_peaks(x, height=2)

plt.plot(x)
plt.plot(peaks, x[peaks], '*r')
plt.hlines(y=2, xmin=0, xmax=len(x), linestyles='dashed', colors='orange')
plt.title("Detecção de picos")
```

```python
# Integração de uma função

from scipy.integrate import quad

def func_int(x, a, b):
    return a*np.cos(x+b)

a = 2
b = np.pi

I = quad(func_int, 0, 2*np.pi, args=(a,b))

print(f"Resultado da integração: {I[0]:.4} / Erro absoluto: {I[-1]:.4}")
```

```python
# Integração de um sinal

from scipy.integrate import cumulative_trapezoid

t = np.linspace(0, 10, 100)
y = 0.25*t

integral = cumulative_trapezoid(y, t, initial=0)

plt.plot(t,y, label="Sinal original")
plt.plot(t, integral, label="Sinal integrado pelo Scipy", linewidth=5)
y2 = 0.25*(t**2)/2
plt.plot(t, y2, label="Sinal integrado real")
plt.legend()
```

```python
from scipy.misc import derivative

def f(t):
    return np.sin(t)

t = np.linspace(0, 2*np.pi, 100)
dy_dt = [derivative(f, tempo, dx=1e-6) for tempo in t] #cria uma lista

plt.plot(t, f(t), label="Sinal")
plt.plot(t, dy_dt, label="Derivada derivada do sinal")
plt.xlabel("Tempo (s)")
plt.ylabel("Amplitude")
plt.legend(loc="lower left")
plt.grid()
```

Resolver o problema de valor inicial: $\frac{dy}{dt} = -2y$, com $y(0)=1$


```python
# Resolvendo PVI de uma EDO

from scipy.integrate import solve_ivp

def edo(t, y):
    return -2 * y

intervalo_tempo = (0, 5)
y0 = [1]
t = np.linspace(0, 5, 100)

sol = solve_ivp(edo, intervalo_tempo, y0, t_eval=t)

plt.plot(sol.t, sol.y[0])
plt.xlabel("Tempo")
plt.title("Solução da EDO")
plt.grid()
plt.show()
```
